```python
import pandas as pd

# üìò DATAFRAME SETUP
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 29],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Salary': [50000, 60000, 70000, 80000, 62000]
}

df = pd.DataFrame(data)

# üîç Preview the data
print(df)

# üëâ TODOs (Quiz-style)

# 1. Display the first 3 rows of the DataFrame
print(df.head(3))

# 2. What are the column names?
print(df.columns)

# 3. What is the average salary?
print(df['Salary'].mean)

# 4. Add a new column called "Seniority" where:
#    - 'Junior' if Age < 30
#    - 'Mid' if Age between 30 and 39
#    - 'Senior' if Age >= 40
df['Seniority'] = df['Age'].apply(
    lambda age: 'Junior' if age < 30 else ('Mid' if age <= 39 else 'Senior')
)

# 5. Filter and show only the employees from the 'IT' department
print(df[df['Department']=='IT'])

# 6. What is the maximum salary in the HR department?
print(df[df['Department']=='HR']['Salary'].max)

# 7. Change the salary of 'Charlie' to 75000
df.loc[df['Name']=='Charlie','Salary']=75000
print(df)
# 8. Remove the row where Name is 'David'
df=df.drop(df.loc[df['Name']=='David'].index,axis=0)
print(df)

# 9. Sort the DataFrame by Salary in descending order
df=df.sort_values(by='Salary',axis=0,ascending=False)
print(df)

# 10. Group by Department and calculate average Age and Salary for each group
df=df.groupby('Department')[['Age','Salary']].mean()
print(df)

# Bonus:
# 11. Export the modified DataFrame to a CSV file called `employee_summary.csv`
df.to_csv('output.csv',index=False)
```
